**Branching**

Branching allows for differnet statements to run for different inputs.

Think of an if statement like a locked room. If the statement is True, we can enter and the program will run. If the statement is False, the program will ignore the task.

To check conditions in the if statement, use the *if condition statement:*, which coontains a word *if*, any condition statement, and a colon at the end. Start the tasks that need to be executed under this condition in a new line with an indent. The lines of code trhat follow the colon and with an indent will only be executed when the *if statemen* is True. The tasks will end when the line of code does not contain the indent.

```python
#If statement

age = 23
#age = 22

#expression that can be true or false
if age > 22:

    #within an indent, the expression that is run if the condition is true
    print("can enter")
    
#the statements after the if statement will run regardless if the statement is true or false
print("move on")
```

The *else* statement runs a block of code if none of the conditions are True before this *else* statement. The code in the *else* block will only be executed if the result is False.

The syntax of the *else* statement is similar to the syntax of the *if* statement.

```python
#else statement

age = 21
#age = 22

if age > 21:
    print("can enter")
else:
    print("cannot enter")
    
print("move on")
```

The *elif* statement is short for else if. This allows us to check additional conditions if the condition statements before it are False.

If the condition for the *elif* statement is True, the alternate expressions will be run.

The syntax of the *elif* statement is similar in that we merely change the *if* in the *if* statement to *elif*.

```python
#elif statement

age = 21

if age > 21:
    print("can enter")
elif age == 21:
    print("go to bar")
else:
    print("cannot enter")
    
print("move on")
```

**Scope of a Variable**

The scope of a variable is the part of that program where that variable is accessible. Variables declared outside of all function definitions are accessible from anywhere within the program. As a result, such variables are said to have global scope, and are called global variables. See the function below:

```python
#global variable

cellphone = "Apple"

def cellphone_brand_rating(brand):
    if brand == cellphone:
        return 10.0
    else:
        return 0.0
    
print("Apple's IPhone rating is:", cellphone_brand_rating("Apple"))
print("Samsung's phone rating is:", cellphone_brand_rating("Samsung"))
print("My favorite cellphone brand is:", cellphone)
```

As shown below, the variable *cellphone* is now defined within the *cellphone_brand_rating* function. The *cellphone_brand_rating* function will still work, because *cellphone* is still defined within the function. However, we can no longer print *cellphone* outside of the function, because it is a local variable of *cellphone_brand_rating*. It is only defined within the *cellphone_brand_rating* function.

A variable that is defined within a function is called a local variable of that function. It is only accessible from within the function in which it is defined.

```python
#deleting variable *cellphone* from example above to demonstrate the local variable

del cellphone

#using local variable

def cellphone_brand_rating(brand):
    cellphone = "Apple"
    if brand == cellphone:
        return 10.0
    else:
        return 0.0
    
print("Apple's IPhone rating is:", cellphone_brand_rating("Apple"))
print("Samsung's phone rating is:", cellphone_brand_rating("Samsung"))
print("My favorite cellphone brand is:", cellphone)
```

As shown below, there are two *cellphone* variable definitions. First one has a global scope, and the second is a local variable within the *cellphone_brand_rating* function.  Within the *cellphone_brand_rating* function, the local variable takes precedence. Samsung will receive a rating of 10.0 when passed to the *cellphone_brand_rating* function. However, outside of the *cellphone_brand_rating* function, the *cellphone_brand_rating* local variable is not defined, so the variable *cellphone* is the global variable, which has a value of Apple.

```python
#global variable and local variable with the same name

cellphone = "Apple"

def cellphone_brand_rating(brand):
    cellphone = "Samsung"
    if brand == cellphone:
        return 10.0
    else:
        return 0.0
    
print("Apple's IPhone rating is:", cellphone_brand_rating("Apple"))
print("Samsung's phone rating is:", cellphone_brand_rating("Samsung"))
print("My favorite cellphone brand is:", cellphone)
```

**Creating a Class**

First, import a library to draw the objects.

```python
# Import the library

import matplotlib.pyplot as plt
%matplotlib inline  
```

First step in creating a class is to use the *class* keyword, then the name of the class.

See image below as a reference.

The next step is a special method called a constructor *_init_*. This is used to initalize the object. The inputs are data attributes. The term *self* contains all the attributes in the set.

For example: the *self.color* gives the value of the attribute color and *self.radius* will give the radius of the object. With the method *add_radius( )* with the parameter *r*, the method adds the value of r to the attribute radius. To access the radius, use the syntax *self.radius*.

Use the method *drawCircle* to display the image of the circle.

```python
# Create a class: Circle

class Circle(object):
    
    # Constructor
    def __init__(self, radius=3, color='blue'):
        self.radius = radius
        self.color = color 
    
    # Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)
    
    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()  
```

**Linspace**

A useful function for plotting mathematical functions is *linspace*. *Linspace* returns evenly spaced numbers over a specified interval. The starting point and ending point of the sequence is specified. The parameter "num" indicates the number of samples to generate.

```python
#makeup a numpy array within [-1, 2] and has 6 elements

np.linspace(-1, 2, num=6)
```

```python
#change the parameter num to 3

np.linspace(-1, 2, num=3)
```

```python
#make numpy array within [0, 2pi] and 100 elements

x = np.linspace(0, 2*np.pi, num=100)
```

```python
#apply sine function to each element in the array x and assign it to the array y

y = np.sin(x)
```

```python
#plot result

plt.plot(x, y)
```

Two Dimensional Numpy

```python
#import the libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
#create a list (this list contains three nested lists each of equal size)

a = [[1, 2, 3], [10, 20, 30], [25, 35, 45]]
a
```

Cast the list to a Numpy Array.

```python
#convert list to numpy array
#every element is the same type

A = np.array(a)
A
```

The attribute *ndim* can use used to obtain the number of axes or dimensions, referred to as the rank.

```python
#show numpy array dimensions

A.ndim
```

The attribute *shape* returns a tuple corresponding to the size or number of each dimension.

```python
#show numpy array shape

A.shape
```

The attribute size gives the total number of elements in the array.

```python
#show numpy array size

A.size
```

**Assessing Different Elements of a Numpy Array**

Rectangular brackets are used to access the different elements of the array. The correspondence between the rectangular brackets and the list and the rectangular representation is shown in the figure below as an example.

Back to the numpy array that was created, A.

```python
#access te element on the second row and second column in A

A[1,1]
```

```python
#access the element in the first row and third column

A[0][2]
```

Splicing can also be used in numpy arrays.

```python
#access the element on the first row and first and second columns

A[0][0:2]
```

```python
#access the element on the second and third rows and first column

A[1:3, 0]
```

**Basic Operations**

Arrays can be added. The process is identical to matrix addition. Matrix addition of X and Y is shown in the image below.

```python
#create numpy array X

X = np.array([[1, 0], [0, 1]])
X
```

```python
#create numpy array Y

Y = np.array([[2, 1], [1, 2]])
Y
```

```python
#add numpy array X and Y

Z = X + Y
Z
```

Multiplying a numpy array by a scale is identical to multiplying a matrix by a scaler. By multiplying the matrix *Y* by the scaler 2, each element in the matrix is multiplied by 2, as shown below.

```python
#create numpy array Y

Y = np.array([[2,3], [1,2]])
Y
```

```python
#multiply array Y by 2

Z = 2 * Y
Z
```

Multiplication of two arrays corresponds to an element-wise product or *Hadamard product*. The *Hadamard product* corresponds to multiplying each of the elements in the same position.

As shown in the figue below, the elements contained in the same color boxes are multiplied. The result is a new matrix that is the same size as matrix *X* or *Y*.

```python
#create numpy array X

X = np.array([[1, 0], [0, 1]])
X
```

```python
#create numpy array Y

Y = np.array([[2, 1], [1, 2]])
Y
```

```python
#multiply X and Y

Z = X * Y
Z
```

Matrix multipilcation can also be used with numpy arrays.

```python
#create matrix A

D = np.array([[-1, 3, 5], [1, 0, 5]])
D
```

```python
#create matrix B

E = np.array([[1, 4], [3, 7], [-1, 6]])
E
```

The numpy function *dot* is used to multiply the arrays together.

```python
#calculate the dot product

F = np.dot(D,E)
F
```

```python
#calculate sine of F

np.sin(F)
```

The numpy attribute *T* can be used to calculate the transposed matrix.

```python
#create matrix R

R = np.array([[1, 2], [4, 5], [8, 9]])
R
```

```python
#get the transposed of R

R.T
```
