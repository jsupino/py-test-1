**Branching**

Branching allows for differnet statements to run for different inputs.

Think of an if statement like a locked room. If the statement is True, we can enter and the program will run. If the statement is False, the program will ignore the task.

To check conditions in the if statement, use the *if condition statement:*, which coontains a word *if*, any condition statement, and a colon at the end. Start the tasks that need to be executed under this condition in a new line with an indent. The lines of code trhat follow the colon and with an indent will only be executed when the *if statemen* is True. The tasks will end when the line of code does not contain the indent.

```python
#If statement

age = 23
#age = 22

#expression that can be true or false
if age > 22:

    #within an indent, the expression that is run if the condition is true
    print("can enter")
    
#the statements after the if statement will run regardless if the statement is true or false
print("move on")
```

The *else* statement runs a block of code if none of the conditions are True before this *else* statement. The code in the *else* block will only be executed if the result is False.

The syntax of the *else* statement is similar to the syntax of the *if* statement.

```python
#else statement

age = 21
#age = 22

if age > 21:
    print("can enter")
else:
    print("cannot enter")
    
print("move on")
```

The *elif* statement is short for else if. This allows us to check additional conditions if the condition statements before it are False.

If the condition for the *elif* statement is True, the alternate expressions will be run.

The syntax of the *elif* statement is similar in that we merely change the *if* in the *if* statement to *elif*.

```python
#elif statement

age = 21

if age > 21:
    print("can enter")
elif age == 21:
    print("go to bar")
else:
    print("cannot enter")
    
print("move on")
```

The *for* loop enables the execution of a code block multiple times. It is used for a controlled flow of repetition.

The code in the indent is executed *N* times. Each time, the value of *i* is increased by 1 for every execution. The statement executed is to *print* the value in the list at the index *i*.

```python
#for loop

dates = [1990, 2022, 2002, 2000]
N = len(dates)

for i in range(N):
    print(dates[i])
```

```python
#for loop: print numbers 0-10

for i in range(0,11):
    print(i)
```

To diectly access elements in a list:

```python
#for loop, loop through list

for year in dates:
    print(year)
```

Changing elements in a list.

```python
#for loop to change elements

circles = ['black', 'red', 'green', 'yellow', 'pink']

for i in range(0,5):
    print("Before Circle", i, 'is', circles[i])
    circles[i] = 'blue'
    print("After Circle", i, 'is', circles[i])
```

Access the index and elements in a list.

```python
#for loop through list and iterate both index and element value

circles = ['black', 'red', 'green', 'yellow', 'pink']

for i, circle in enumerate(circles):
    print(i, circle)
```

*While* Loop

The *while* loop exists as a tool for repeated execution based on a conditon. The code block will be exectuted until the given logical condition returns a False Boolean value.

```python
#while loop

dates = [1990, 2022, 2002, 2000]

i = 0
year = dates[0]

while(year !=2022):
    print(year)
    i = i + 1
    year = dates[i]
    
print("it took", i, "repetitions to get out of loop.")
```
